```python
import pandas
```

```python
obesity=pandas.read_csv('Obesity Classification.csv')
obesity
```

```python
obesity.isnull().sum()
```

```python
obesity['Gender'].replace({
    'Male':0,
    'Female':1
},inplace=True)
obesity.groupby('Gender').size()
```

```python
array=obesity.values
X=array[:,1:6]
X
Y=array[:,6]
Y
```

```python
from sklearn import model_selection
x_train,x_test,y_train,y_test=model_selection.train_test_split(X,Y,test_size=0.3,random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
prediction=model.predict(x_test)
print('Predictions are',prediction)
print(' actual answers are',y_test)
```

```python
from sklearn.metrics import accuracy_score
check_accuracy=accuracy_score(prediction,y_test)*100
check_accuracy
```
