```python
import pandas
```

```python
customer=pandas.read_csv('customerdata.csv')
customer
```

```python
customer.isnull().sum()
```

```python
array=customer.values
X=array[:,2:7]
X
```

```python
from sklearn.cluster import KMeans as km
model=km(n_clusters=15,random_state=42)
model.fit(X)
```

```python
groups=model.cluster_centers_
groups
```

```python
clusters=pandas.DataFrame(groups,columns=['AAvg_Credit_Limit','Total_Credit_Cards','Total_visits_bank','Total_visits_online','Total_calls_made'])
clusters
```
